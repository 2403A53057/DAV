# Task
Analyze patient treatment and outcome data from "/content/Patient.csv", "/content/Treatment.csv", and "/content/outcomes.csv" by merging the datasets, grouping by treatment type to find average recovery days and success rate, and pivoting the data to show each patient's total cost per treatment type.

## Load data

### Subtask:
Load the three provided CSV files (`Patient.csv`, `Treatment.csv`, and `outcomes.csv`) into pandas DataFrames.


**Reasoning**:
Load the three CSV files into pandas DataFrames as instructed.



```python
import pandas as pd

df_patient = pd.read_csv('/content/Patient.csv')
df_treatment = pd.read_csv('/content/Treatment.csv')
df_outcomes = pd.read_csv('/content/outcomes.csv')

display(df_patient.head())
display(df_treatment.head())
display(df_outcomes.head())
```

## Merge data

### Subtask:
Merge the three DataFrames based on relevant columns (likely patient ID).


**Reasoning**:
Merge the three dataframes based on the relevant columns and display the head of the merged dataframe.



```python
df_merged = pd.merge(df_patient, df_treatment, on='Patient_id')
df_merged = pd.merge(df_merged, df_outcomes, on='Treatment_id')
display(df_merged.head())
```

## Calculate aggregates

### Subtask:
Group the merged data by `treatment_type` and calculate the average `recovery_days` and the `success_rate`.


**Reasoning**:
Group the merged data by treatment type and calculate the average recovery days and success rate.



```python
df_grouped = df_merged.groupby('Treatment_type').agg(
    average_recovery_days=('Recovery_days', 'mean'),
    success_rate=('Success', lambda x: (x == 'yes').mean())
)
display(df_grouped)
```

## Pivot data

### Subtask:
Pivot the merged data to display each patient's total cost per `treatment_type`.


**Reasoning**:
Pivot the merged dataframe to show the total cost per treatment type for each patient and display the result.



```python
df_pivot = df_merged.pivot_table(index='Patient_id', columns='Treatment_type', values='Cost', aggfunc='sum')
display(df_pivot.head())
```

## Display results

### Subtask:
Display the calculated aggregates and the pivoted table.


**Reasoning**:
Display the grouped and pivoted DataFrames as requested by the subtask.



```python
display(df_grouped)
display(df_pivot)
```
