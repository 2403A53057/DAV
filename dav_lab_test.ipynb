```python
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Dataset
data = {
    "product": ["electronics", "books", "accessories", "food"],
    "sales": [900, 420, 815, 1500],
    "profit": [91, 51, 71, 95]
}

df = pd.DataFrame(data)

app = dash.Dash(__name__)

# Layout
app.layout = html.Div([

    html.H2("Product Performance Dashboard"),

    dcc.Dropdown(
        id='metric',
        options=[
            {'label': 'Sales', 'value': 'sales'},
            {'label': 'Profit', 'value': 'profit'}
        ],
        value='sales',
        clearable=False
    ),

    dcc.Graph(id='line_chart')
])

# Callback
@app.callback(
    Output('line_chart', 'figure'),
    Input('metric', 'value')
)
def update_graph(selected_metric):

    fig = px.line(
        df,
        x="product",
        y=selected_metric,
        title=f"{selected_metric.capitalize()} over Products",
        markers=True
    )

    return fig

# Run app
if __name__ == '__main__':
    app.run(debug=True)
```
